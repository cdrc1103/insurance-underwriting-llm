```python
from pathlib import Path
from datasets import load_from_disk

dataset = load_from_disk(str(Path.cwd().parent / "data" / "splits" / "test"))
print(f"Loaded {len(dataset)} examples")
print(f"Features: {list(dataset.features.keys())}")
```

```python
def show_example(idx: int):
    """Display example in readable format."""
    ex = dataset[idx]
    print(f"{'=' * 80}")
    print(f"Example {idx} | Task: {ex['task']} | Turns: {ex['num_turns']}")
    print(f"{'=' * 80}\n")

    for msg in ex["messages"]:
        role = msg["role"].upper()
        content = msg["content"]
        print(f"[{role}]")
        print(content)
        print()
```

```python
# Example with <think> tags (internal reasoning)
show_example(0)
```
